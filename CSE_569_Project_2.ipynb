# Step 1: Feature combination by fusion of classifiers. 

```python
sample = [0 for i in range(0, len(preds1))]
correct = incorrect = 0
for j in range(0, len(preds1)):
  maxindex = 0
  maxarg = -1
  for i in range(0, len(preds1[0])):
    mean = (preds1[j][i] + preds2[j][i] + preds3[j][i])/3
    if(maxarg < mean):
      maxarg = mean
      maxindex = i + 1
    
  sample[j] = maxindex

for k in range(0, len(preds1)):
  if sample[k] == npy_test[k]:
    correct += 1
  else:
    incorrect += 1

[correct/len(preds1), correct, incorrect]
```

# Step 2: Feature combination by simple concatenation

```python
concat_train = np.concatenate((train_data['X1'], train_data['X2'],train_data['X3']), axis=1)
concat_test = np.concatenate((test_data['X1'], test_data['X2'],test_data['X3']), axis=1)
```

```python
prob_s2 = svm_problem(npy.flatten(), concat_train)
step2_train = svm_train(prob_s2, param)
```

```python
p_label_s2, p_acc_s2, preds_s2 = svm_predict(test_data['Y'].flatten(), concat_test, step2_train)
```
